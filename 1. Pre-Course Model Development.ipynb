```python
# IMPORTING THE DATA

# Importing the pandas package 
import pandas as pd

# Loading Apprentice Chef dataset
file = "./Apprentice_Chef_Dataset.xlsx"
apprentice = pd.read_excel(io = file)
```

<h3> MODEL N°1: OLS </h3>

```python
# Importing the statsmodels package
import statsmodels.formula.api as smf
```

```python
# Creating a regression model with all numeric variables
lm_revenue = smf.ols(formula = 'REVENUE ~ CROSS_SELL_SUCCESS + TOTAL_MEALS_ORDERED + UNIQUE_MEALS_PURCH + CONTACTS_W_CUSTOMER_SERVICE + PRODUCT_CATEGORIES_VIEWED + AVG_TIME_PER_SITE_VISIT + MOBILE_NUMBER + CANCELLATIONS_BEFORE_NOON + CANCELLATIONS_AFTER_NOON + TASTES_AND_PREFERENCES + PC_LOGINS + MOBILE_LOGINS + WEEKLY_PLAN + EARLY_DELIVERIES + LATE_DELIVERIES + PACKAGE_LOCKER + REFRIGERATED_LOCKER + AVG_PREP_VID_TIME + LARGEST_ORDER_SIZE + MASTER_CLASSES_ATTENDED + MEDIAN_MEAL_RATING + AVG_CLICKS_PER_VISIT + TOTAL_PHOTOS_VIEWED',
             data = apprentice)

# Showing summary of regression
print(lm_revenue.fit().summary())


## Numerous features have a high p-value (>0.05) and are not statistically 
## significant. Features with high p-values will be removed one by one to 
## improve the model until all features are statistically significant. 
```

```python
# Creating a regression model with only statistically significant variables
lm_revenue = smf.ols(formula = 'REVENUE ~ CROSS_SELL_SUCCESS + TOTAL_MEALS_ORDERED + UNIQUE_MEALS_PURCH + CONTACTS_W_CUSTOMER_SERVICE + AVG_TIME_PER_SITE_VISIT + AVG_PREP_VID_TIME + LARGEST_ORDER_SIZE + MASTER_CLASSES_ATTENDED + MEDIAN_MEAL_RATING + TOTAL_PHOTOS_VIEWED',
             data = apprentice)

# Showing summary of regression
print(lm_revenue.fit().summary())

## The adjusted r-square value is the same compared to the one from the 
## previous model. The new model has ten features, meaning that all 
## unsignificant features have been removed. All features remaining are 
## statistically significant (p-value < 0.05). The adjusted R-squared is 
## 0.636, which means that 64% of the revenue is explained by the ten 
## variables.
```

<br>
<h3> MODEL N°2: LINEAR REGRESSION USING SCIKIT-LEARN </h3>

```python
# Importing necessary packages
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Creating a model with all numeric variables

# Creating features and target variable
X = apprentice.loc[:,['CROSS_SELL_SUCCESS', 'TOTAL_MEALS_ORDERED','UNIQUE_MEALS_PURCH', 'CONTACTS_W_CUSTOMER_SERVICE', 'PRODUCT_CATEGORIES_VIEWED', 'AVG_TIME_PER_SITE_VISIT', 'MOBILE_NUMBER', 'CANCELLATIONS_BEFORE_NOON','CANCELLATIONS_AFTER_NOON','TASTES_AND_PREFERENCES', 'PC_LOGINS', 'MOBILE_LOGINS','WEEKLY_PLAN', 'EARLY_DELIVERIES', 'LATE_DELIVERIES', 'PACKAGE_LOCKER', 'REFRIGERATED_LOCKER', 'AVG_PREP_VID_TIME', 'LARGEST_ORDER_SIZE', 'MASTER_CLASSES_ATTENDED',  'MEDIAN_MEAL_RATING', 'AVG_CLICKS_PER_VISIT', 'TOTAL_PHOTOS_VIEWED']]
y = apprentice.loc[:,['REVENUE']]

# Splitting the data in training and testing set
X_train, X_test, y_train, y_test = train_test_split(X,y)

# Creating the regression
lr = LinearRegression()

# Fitting the regression model to the training data
lr_fit = lr.fit(X_train, y_train)

# Predicting on the test data
lr_prediction = lr_fit.predict(X_test)

# Printing model score
print('R-squared training set:', lr.score(X_train, y_train).round(3))  
print('R-squared testing set:',  lr.score(X_test, y_test).round(3)) 


## With all numeric variables, the adjusted R-square is 0.64 for the training 
## set and 0.638 for the testing set. Those results are really similar to
## the first regression model (OLS).
```

```python
# Creating a model with only statistically significant variables

# Creating features and target variable
X = apprentice.loc[:,['CROSS_SELL_SUCCESS', 'TOTAL_MEALS_ORDERED', 'UNIQUE_MEALS_PURCH', 'CONTACTS_W_CUSTOMER_SERVICE', 'AVG_TIME_PER_SITE_VISIT', 'AVG_PREP_VID_TIME', 'LARGEST_ORDER_SIZE', 'MASTER_CLASSES_ATTENDED', 'MEDIAN_MEAL_RATING','TOTAL_PHOTOS_VIEWED']]
y = apprentice.loc[:,['REVENUE']]

# Splitting the data in training and testing set
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 21)

# Creating the regression
lr = LinearRegression()

# Fitting the model to the training data
lr_fit = lr.fit(X_train, y_train)

# Predicting on the test data
lr_prediction = lr_fit.predict(X_test)

# Printing model score
print('R-squared training set:', lr.score(X_train, y_train).round(3))  
print('R-squared testing set:',  lr.score(X_test, y_test).round(3)) 

## With exclusively the ten significant variables identified in the OLS 
## regression, the adjusted R-square is 0.633 for the training set and 
## 0.653 for the testing set. Again,the results are really similar to 
## the model build previously.
```
